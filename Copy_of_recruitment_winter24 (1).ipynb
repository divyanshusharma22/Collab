# 3. Initialize the OCR model
###Initialize the EasyOCR reader
###Use easyocr.Reader() with the desired language(s)

```python
reader = easyocr.Reader(['en'])
print("EasyOCR model initialized successfully!")
```

# 4. Perform text recognition
#### TODO: Use the EasyOCR reader to extract text from the image
#### Hint: Use the .readtext() method

```python
# Recognise the text
# Perform OCR on the image
results = reader.readtext(img_path)  # Pass the image path to the .readtext() method

# Display the extracted text and confidence scores
for idx, (bbox, text, prob) in enumerate(results):
    print(f"Result {idx + 1}:")
    print(f"  Text: {text}")
    print(f"  Confidence: {prob:.2f}")
    print("-" * 30)

```

# 5. Process and display the results
#### TODO: Print or format the detected text along with its confidence score

```python
#compute the result
print("Extracted Text and Confidence Scores:")
for idx, (bbox, text, prob) in enumerate(results):
    print(f"Result {idx + 1}:")
    print(f"  Text: {text}")       # The detected text
    print(f"  Confidence: {prob:.2f}")  # Confidence score
    print("-" * 30)
```

```python
#show the result
print("\nSummary:")
print(f"Total Text Segments Detected: {len(results)}")
```

# Example Output:
### Detected Text: "Sample Text" (Confidence: 0.98)

```python
# Extracting text from the image


# Initialize the EasyOCR reader for English
reader = easyocr.Reader(['en'])

# Perform OCR on the image
results = reader.readtext(img_path)

# Process and display the extracted text and its confidence score
for idx, (bbox, text, prob) in enumerate(results):
    print(f"Detected Text: \"{text}\" (Confidence: {prob:.2f})")

```

# Save or display the final result
###Save the processed image or display it using cv2 or plt

```python
# Extracted Text

# Load the selected image
image = cv2.imread(img_path)

# Perform OCR on the image
results = reader.readtext(img_path)

# Draw bounding boxes and overlay recognized text on the image
for bbox, text, prob in results:
    # Get the top-left and bottom-right corners of the bounding box
    top_left = tuple(map(int, bbox[0]))
    bottom_right = tuple(map(int, bbox[2]))

    # Draw the bounding box around the detected text
    image = cv2.rectangle(image, top_left, bottom_right, (0, 255, 0), 2)  # Green box

    # Overlay the detected text on the image
    image = cv2.putText(image, f"{text} ({prob:.2f})", top_left, cv2.FONT_HERSHEY_SIMPLEX,
                        0.4, (255, 0, 0), 1)  # Blue text with confidence score

# Display the image with the bounding boxes and text
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Image with Recognized Text")
plt.axis('off')
plt.show()

```
